```python
# Unique values of all features (for more information, please see the links above):

# age: continuous.
# workclass: Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, Local-gov, State-gov, Without-pay, Never-worked.
# fnlwgt: continuous.
# education: Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc,
#            9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th, Preschool.
# education-num: continuous.
# marital-status: Married-civ-spouse, Divorced, Never-married, Separated, Widowed,
#                 Married-spouse-absent, Married-AF-spouse.
# occupation: Tech-support, Craft-repair, Other-service, Sales, Exec-managerial,
#             Prof-specialty, Handlers-cleaners, Machine-op-inspct, Adm-clerical,
#             Farming-fishing, Transport-moving, Priv-house-serv, Protective-serv, Armed-Forces.
# relationship: Wife, Own-child, Husband, Not-in-family, Other-relative, Unmarried.
# race: White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black.
# sex: Female, Male.
# capital-gain: continuous.
# capital-loss: continuous.
# hours-per-week: continuous.
# native-country: United-States, Cambodia, England, Puerto-Rico, Canada, Germany,
#                 Outlying-US(Guam-USVI-etc), India, Japan, Greece, South, China,
#                 Cuba, Iran, Honduras, Philippines, Italy, Poland, Jamaica, Vietnam,
#                 Mexico, Portugal, Ireland, France, Dominican-Republic, Laos, Ecuador,
#                 Taiwan, Haiti, Columbia, Hungary, Guatemala, Nicaragua, Scotland, Thailand,
#                 Yugoslavia, El-Salvador, Trinadad&Tobago, Peru, Hong, Holand-Netherlands.
# salary: >50K,<=50K
```

```python
import pandas as pd

data = pd.read_csv("adult.data.csv")
data.head()
```

```python
data.columns
```

```python
# How many men and women (sex feature) are represented in this dataset?

data['sex'].value_counts()
```

```python
# What is the average age (age feature) of women?

data[data['sex'] == 'Female'].mean()['age']
```

```python
# What is the percentage of German citizens (native-country feature)?

round((data[data['native-country'] == 'Germany'].shape[0] / data.shape[0]) * 100, 2)
```

```python
# What are the mean and standard deviation of age
# for those who earn more than 50K per year (salary feature)
# and those who earn less than 50K per year?

fitty_gt = data[data['salary'] == '>50K']['age']
fitty_lt = data[data['salary'] == '<=50K']['age']

answer = f"{fitty_gt.mean()=}, {fitty_gt.std()=};\n {fitty_lt.mean()=}, {fitty_lt.std()=}"
answer
```

```python
# Is it true that people who earn more than 50K have at least high school education?
# (education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate feature)

educations = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
data[(data['salary'] == '>50K') & (data['education'].apply(lambda x: x not in educations))]
```

```python
# Display age statistics for each race (race feature) and each gender (sex feature).
# Use groupby() and describe(). Find the maximum age of men of Amer-Indian-Eskimo race.

demo_data = data.groupby(['race', 'sex'])['age']
demo_data_describ = demo_data.describe()
demo_data_describ
```

```python
demo_data_describ.loc['Amer-Indian-Eskimo', 'Male'].loc['max']
```

```python
# Among whom is the proportion of those who earn a lot (>50K) greater:
# married or single men (marital-status feature)? Consider as married those who have a marital-status
# starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse),
# the rest are considered bachelors.
maritals = ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']
gt_fifty = data[(data['salary'] == '>50K') & (data['sex'] == 'Male')]
all_riches = gt_fifty.shape[0]
married_riches = gt_fifty[gt_fifty['marital-status'].apply(lambda x: x in maritals)].shape[0]
unmarried_riches = all_riches - married_riches

f"{married_riches=}, {unmarried_riches=}"
```

```python
# What is the maximum number of hours a person works per week (hours-per-week feature)?
# How many people work such a number of hours,
# and what is the percentage of those who earn a lot (>50K) among them?

max_hpw = data['hours-per-week'].max()
hard_workers_raw = data[data['hours-per-week'] == max_hpw]
hard_workers = hard_workers_raw.shape[0]
rich_hard_work = (hard_workers_raw[hard_workers_raw['salary'] == '>50K']).shape[0]
rich_to_pure = round(rich_hard_work / hard_workers * 100, 2)

f"{max_hpw=}, {hard_workers=}, {rich_to_pure=}%"
```

```python
# Count the average time of work (hours-per-week) for those who earn a little and
# a lot (salary) for each country (native-country). What will these be for Japan?

gt=(data[data['salary'] == '>50K']).groupby(['native-country'])['hours-per-week'].describe()
round(gt['mean'].loc['Japan'], 2)
```

```python
lt=(data[data['salary'] != '>50K']).groupby(['native-country'])['hours-per-week'].describe()
lt['mean'].loc['Japan']
```
